```python
import numpy as np
import matplotlib.pyplot as plt

def get_waveform(img):
    # convert 2D image to 1D signal
    waveform = np.mean(img, axis=0)   # average along height
    return waveform
```

```python
plt.figure(figsize=(12, 8))

# Normal
plt.subplot(4, 1, 1)
plt.plot(get_waveform(normal_image[0]))
plt.title("Normal ECG Waveform")

# MI
plt.subplot(4, 1, 2)
plt.plot(get_waveform(mi_image[0]))
plt.title("MI ECG Waveform")

# History MI
plt.subplot(4, 1, 3)
plt.plot(get_waveform(history_image[0]))
plt.title("History MI ECG Waveform")

# Abnormal
plt.subplot(4, 1, 4)
plt.plot(get_waveform(abnormal_image[0]))
plt.title("Abnormal ECG Waveform")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))

# Normalize each waveform (so comparison is fair)
def normalize(x):
    return (x - x.min()) / (x.max() - x.min())

# Plot with better visibility
plt.plot(normalize(get_waveform(normal_image[0])), linewidth=2, label="Normal")
plt.plot(normalize(get_waveform(mi_image[0])), linewidth=2, linestyle='--', label="MI")
plt.plot(normalize(get_waveform(history_image[0])), linewidth=2, linestyle='-.', label="History MI")
plt.plot(normalize(get_waveform(abnormal_image[0])), linewidth=2, linestyle=':', label="Abnormal")

# Labels & styling
plt.legend(fontsize=12)
plt.title("ECG Waveform Comparison (Normalized)", fontsize=14)
plt.xlabel("Time", fontsize=12)
plt.ylabel("Normalized Amplitude", fontsize=12)

# Grid for clarity
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
image = normal_image + mi_image + history_image + abnormal_image

labels = (
    normal_lables +
    mi_lables +
    history_lables +
    abnormal_lables
)
```

```python
import numpy as np

train_feature = np.array(image)
labels = np.array(labels)

print(f"image dataset shape = {train_feature.shape}")
print(f"label dataset shape = {labels.shape}")
```

```python
# ===============================
# Reshape for CNN (IMPORTANT)
# ===============================
train_feature = train_feature.reshape(-1, 200, 240, 1)
```

```python
# ===============================
# Train-Test Split
# ===============================
from sklearn.model_selection import train_test_split

train_features, test_features, train_target, test_target = train_test_split(
    train_feature, labels, test_size=0.1, random_state=42
)

print(f"train_features shape = {train_features.shape}")
print(f"test_features shape = {test_features.shape}")
print(f"train_target shape = {train_target.shape}")
print(f"test_target shape = {test_target.shape}")
```

```python
# ===============================
# MODEL ARCHITECTURE
# ===============================
import tensorflow as tf
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout

# One-hot encoding
train_target = to_categorical(train_target)
test_target = to_categorical(test_target)

# Model
model = Sequential()

# Convolution Layers


model.add(Conv2D(32, kernel_size=(3,3), padding='same', activation='relu', input_shape=(200, 240, 1)))
model.add(MaxPooling2D(pool_size=(2,2)))  # → (100, 120, 32)

model.add(Conv2D(64, kernel_size=(3,3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))  # → (50, 60, 64)

model.add(Conv2D(64, kernel_size=(3,3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))  # → (25, 30, 64)

model.add(Flatten())  # → 25 × 30 × 64 = 48,000 (manageable!)

model.add(Dense(500, activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(300, activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(200, activation='relu'))
model.add(Dense(4, activation='softmax'))
model.summary()
```

```python
# ===============================
# MODEL COMPILE
# ===============================
from tensorflow.keras.optimizers import Adam

model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# ===============================
# MODEL TRAINING
# ===============================
history = model.fit(
    train_features,
    train_target,
    validation_data=(test_features, test_target),
    epochs=25,
    batch_size=32
   
)
```
